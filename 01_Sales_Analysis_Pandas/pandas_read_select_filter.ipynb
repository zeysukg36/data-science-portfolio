## 2. Building a DataFrame Manually

Before loading real data, let's understand the DataFrame structure  
by creating a small one from a dictionary.

```python
data = {
    "order_id"      : [201, 202, 203, 204, 205],
    "city"          : ["Istanbul", "Ankara", "Istanbul", "Izmir", "Bursa"],
    "category"      : ["Electronics", "Clothing", "Market", "Electronics", "Market"],
    "quantity"      : [1, 2, 5, 1, 3],
    "unit_price"    : [15000, 900, 90, 42000, 150],
    "payment"       : ["Credit Card", "Bank Transfer", "EFT", "Credit Card", "Cash on Delivery"]
}

df = pd.DataFrame(data)
df
```

### 2.1 Basic DataFrame Info

```python
print("Rows / Columns:", df.shape)
print("Columns:", list(df.columns))
df.info()
```

## 3. Column Selection

### 3.1 Single Column

```python
df["city"]
```

### 3.2 Multiple Columns

```python
df[["order_id", "category", "quantity", "unit_price"]]
```

### 3.3 Creating a New Column

```python
df["total_price"] = df["quantity"] * df["unit_price"]
df
```

## 4. Row Selection: `iloc` and `loc`

### 4.1 `iloc` — by index position

```python
df.iloc[0]        # first row
```

```python
df.iloc[0:3]      # first 3 rows
```

### 4.2 `loc` — by label / condition

Our index is 0-based here, so `loc[0]` gives the first row.

```python
df.loc[0]
```

## 5. Filtering (Boolean Mask)


### 5.1 Single city


```python
df[df["city"] == "Istanbul"]
```

### 5.2 Electronics with unit price above 20,000

```python
df[(df["category"] == "Electronics") & (df["unit_price"] > 20000)]
```

### 5.3 Multiple categories with `isin()`

```python
df[df["category"].isin(["Market", "Clothing"])]
```

## 6. Sorting

```python
df.sort_values("total_price", ascending=False)
```

## 7. Quick Summaries: `value_counts` and `describe`

```python
df["category"].value_counts()
```

```python
df[["quantity", "unit_price", "total_price"]].describe()
```

## 8. Loading a Real CSV File

From here we work with a real sales dataset.  
Make sure `ybs_satis_verisi.csv` is in the same folder as this notebook.

```python
sales = pd.read_csv("ybs_satis_verisi.csv")
sales.head()
```

### 8.1 Data Type Check

```python
sales.info()
```

### 8.2 Convert Order Date to datetime

```python
sales["siparis_tarihi"] = pd.to_datetime(sales["siparis_tarihi"])
sales.dtypes
```

## 9. Select Only Relevant Columns

```python
core = sales[["siparis_id", "sehir", "kategori", "urun", "adet", "birim_fiyat"]]
core.head()
```

## 10. Real-World Filtering Scenarios

### 10.1 Electronics orders in Kocaeli

```python
kocaeli_electronics = sales[
    (sales["sehir"] == "Kocaeli") & (sales["kategori"] == "Elektronik")
]
kocaeli_electronics.head()
```

### 10.2 Discount rate 15% and above

```python
high_discount = sales[sales["indirim_orani"] >= 0.15]
high_discount.head()
```

### 10.3 Satisfaction score between 3 and 5 (inclusive)

```python
satisfied = sales[sales["memnuniyet_puani"].between(3, 5)]
satisfied.head()
```

### 10.4 Orders from the last 60 days

```python
cutoff = sales["siparis_tarihi"].max() - pd.Timedelta(days=60)
recent = sales[sales["siparis_tarihi"] >= cutoff]
recent.head()
```

## 11. Sorting + Top 5 Most Expensive Orders

```python
sales.sort_values("birim_fiyat", ascending=False).head(5)[
    ["siparis_id", "sehir", "kategori", "urun", "birim_fiyat", "adet"]
]
```
